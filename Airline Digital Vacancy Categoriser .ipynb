```python
import pandas as pd
import seaborn as sns 
import numpy as np 
from matplotlib import pyplot as plt
import os
import glob
```

```python
df = pd.read_csv('Data_Sources/JB_Glassdoor__1_col_.csv')
```

```python
#categorize the job vacancies on whether they are digital/tech or management using keyworks

searchfor_tech = [' IT ', ' IT', '-IT', 'IT-', 'Digital', 'Technology', 'Automation', 'Developer', ' API ',
                  'Data', 'Agile', '-Entwicklung', ' Entwicklung', '-Entwickler', 'Entwickler', 'Informatik', 
                  'eLearning', 'online', 'Software', 'DevOps', 'Information Management', 'Innovation', 
                  'Product Owner', 'Product Manager', ' Mobile ', 'Architect', ' UX ', 'UX/', ' UX', 'Cisco',' SAP ',
                  'SAP-', 'Scrum', 'Business Intelligence', 'Développement Informatique', 'Transformation',
                  ' Internet ', 'Web ', ' Architecte ', 'E-commerce', 'Tech Ops', ' API', 'API ', 'Hacking',
                  'Analytics', 'IT Analyst']

searchfor_mang = ['Manager', 'Managing Director', 'Head of', 'Leader', 'Director']

df['Technology'] = np.where(df['Column 1'].str.contains('|'.join(searchfor_tech), case=False, na=False), '1', '0')

df['Management'] = np.where(df['Column 1'].str.contains('|'.join(searchfor_mang), case=False, na=False), '1', '0')
```

```python
# share of vacancies that are management 

df.Management.value_counts(normalize=True, sort=True, bins=None, dropna=True)
```

```python
# share of vacancies that are technology

df.Technology.value_counts(normalize=True, sort=True, bins=None, dropna=True)
```

```python
df.sample(n=50)
```

```python
# create a new column Tech_Mang which indicates jobs that are BOTH technolgy and management 

def tech(df):
    if df['Management'] == '1' and df['Technology'] == '1':
        return '1'
    else:
        return '0'
    
df['Tech_Mang'] = df.apply(tech, axis=1)
df
```

```python
# share of vacancies that are both technology and management

y = df.Tech_Mang.value_counts(normalize=False, sort=True, bins=None, dropna=True)
x = df.Management.value_counts(normalize=False, sort=True, bins=None, dropna=True)

Tech_M = y[1]/x[1]
print(Tech_M)
```
